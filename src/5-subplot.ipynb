### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y_line = x**2
categories = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]


fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y_line, color='blue')
axs[0, 0].set_title('Line Chart: y = x^2')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].grid()


axs[0, 1].bar(categories, values_bar, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart of Categories')
axs[0, 1].set_ylabel('Values')


axs[1, 0].scatter(x_scatter, y_scatter, color='purple', s=100)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid()

axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Programming Language Popularity')


plt.tight_layout()


plt.show()
```
